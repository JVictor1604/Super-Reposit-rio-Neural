```python
import random
from pprint import pprint
from functools import partial
from itertools import product

from funcoes_6 import funcao_objetivo_pop_liga
from funcoes_6 import populacao_cb_liga as cria_populacao
from funcoes_6 import selecao_torneio_max as funcao_selecao
from funcoes_6 import cruzamento_uniforme as funcao_cruzamento
from funcoes_6 import mutacao_troca as funcao_mutacao
from funcoes_6 import calcula_liga
```

```python
LIMITE = 100
NUM_PRECO_ELEMENTO_GRAMA = len(preco_elemento_grama_final)
ORDEM_DOS_PRECO_ELEMENTO_GRAMA = list(sorted(preco_elemento_grama_final.keys()))

TAMANHO_POPULACAO = 10
NUM_GERACOES = 5
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_LIGA = 3
```

```python
funcao_objetivo = partial(funcao_objetivo_pop_liga, 
                          itens=preco_elemento_grama_final, 
                          ordem_dos_itens=ORDEM_DOS_PRECO_ELEMENTO_GRAMA, 
                          limite=LIMITE)
```

```python
populacao = cria_populacao(TAMANHO_POPULACAO, len(preco_elemento_grama_final), TAMANHO_LIGA)


pprint(populacao)
```

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Sele√ß√£o
    fitness = funcao_objetivo(populacao)        
    selecionados = funcao_selecao(populacao, fitness, 3)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Muta√ß√£o
    funcao_mutacao(proxima_geracao, CHANCE_DE_MUTACAO)
    
    # Atualiza√ß√£o do hall da fama
    fitness = funcao_objetivo(proxima_geracao)
        
    maior_fitness = max(fitness)
    indice = fitness.index(maior_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
```

```python
fitness = funcao_objetivo(hall_da_fama)
maior_fitness = max(fitness)
indice = fitness.index(maior_fitness)
melhor_individuo_observado = hall_da_fama[indice]

print()
print("Voc√™ deve pegar os seguintes itens:")

for i in range(NUM_PRECO_ELEMENTO_GRAMA):
    if melhor_individuo_observado[i] == 1:
        print("+", ORDEM_DOS_PRECO_ELEMENTO_GRAMA[i])

print()


peso_total, valor_total = calcula_liga(melhor_individuo_observado, preco_elemento_grama_final, NUM_PRECO_ELEMENTO_GRAMA)

print(
    f"Com isso, sua mochila ter√° o valor de {valor_total:.2f} reais "
    f"e peso de {peso_total:.2f} quilogramas."
)
```
